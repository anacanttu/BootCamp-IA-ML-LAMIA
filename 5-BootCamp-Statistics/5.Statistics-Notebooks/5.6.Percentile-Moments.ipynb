# Percentile

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vals = np.random.normal(0,0.5,10000)
plt.hist(vals, 50)
plt.show()
```

```python
np.percentile(vals, 50)
```

```python
np.percentile(vals,90)
```

```python
np.percentile(vals,20)
```

# Moments: Mean, Variance, Skew, Kurtosis


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
valsvals = np.random.normal(0,0.5,10000)
plt.hist(vals, 50)
plt.show()
```

```python
np.mean(vals)
```

```python
np.var(vals)
```

```python
import scipy.stats as sp
sp.skew(vals)
```

```python
sp.kurtosis(vals)
```
